```python
# First code block is the import libraries  
import numpy as np
import pandas as pd
from sklearn import metrics
from matplotlib import pyplot as plt
```

```python
dataPath = '../spaceship-titanic_rawData/spaceship_train.csv'

df_spaceship = pd.read_csv(dataPath)

df_spaceship.shape
```

```python
df_spaceship.head(10)
```

```python
df_spaceship.isna().sum()
```

```python
# Drop rows that has NaN values on selected columns
df_spaceship = df_spaceship.dropna(subset=['CryoSleep','Age', 'VIP', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck'])
```

```python
df_spaceship.value_counts("Destination")
```

```python
# Export splitted data

file_name = '../preprocess_train_dataset/rmNaNDate_01.csv'

df_spaceship.to_csv(file_name, sep=',', encoding='utf-8', index=False)
```

Handle the dataset for testing dataset

```python
test_dataPath = '../spaceship-titanic_rawData/spaceship_test.csv'

df_test_raw = pd.read_csv(test_dataPath)

df_test_raw.shape
```

```python
# We cannot drop the na data in testing dataset
# df_nonNaN_test = df_test_raw.dropna(subset=['CryoSleep','Age', 'VIP', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck'])
df_test_raw = df_test_raw.drop(columns=['PassengerId', 'HomePlanet','Cabin', 'Destination', 'Name'])
# df_nonNaN_test = df_test_raw.fillna(df_test_raw.mean())
```

```python
df_nonNaN_test = df_test_raw.fillna({
    "CryoSleep": df_test_raw["CryoSleep"].mode()[0],
    "Age": df_test_raw["Age"].mean(),
    "VIP": df_test_raw["VIP"].mode()[0],
    "RoomService": df_test_raw["RoomService"].mean(),
    "FoodCourt": df_test_raw["FoodCourt"].mean(),
    "ShoppingMall": df_test_raw["ShoppingMall"].mean(),
    "Spa": df_test_raw["Spa"].mean(),
    "VRDeck": df_test_raw["VRDeck"].mean(),
})
```

```python
df_nonNaN_test
```

```python
# Export splitted data

file_name = '../preprocess_test_dataset/rmNaN_testDate_01.csv'

df_nonNaN_test.to_csv(file_name, sep=',', encoding='utf-8', index=False)
```
